```python
import tensorflow as tf
import libspn as spn
```

# Simple Network

```python
v1=spn.ContVars(num_vars=2)
v2=spn.IVs(num_vars=1, num_vals=2)
p1=spn.Product(v1, v2, (v1, [0]))
p2=spn.Product(v1, v2)
root=spn.Sum(p1, p2, p1)
spn.generate_weights(root)
root.generate_ivs()
```

```python
spn.display_spn_graph(root, skip_params=False)
```

```python
spn.display_spn_graph(root, skip_params=True)
```

# Generated Dense SPN

```python
ivs=spn.IVs(3,3)
gen=spn.DenseSPNGenerator(num_mixtures=2, num_decomps=2, num_subsets=2)
root=gen.generate(ivs)
spn.generate_weights(root)
root.generate_ivs()
```

```python
spn.display_spn_graph(root, skip_params=False)
```

```python
spn.display_spn_graph(root, skip_params=True)
```
