```python
import tensorflow as tf
import libspn as spn
```

```python
components1 = [spn.GaussianMixtureDataset.Component(0.4, [2,4], [[1,0],[0,1]], 0),              
               spn.GaussianMixtureDataset.Component(0.3, [1,1], [[1,0.5],[0.5,1]], 1),
               spn.GaussianMixtureDataset.Component(0.3, [3,1], [[1,0.4],[0.8,1]], 1)]
components2 = [spn.GaussianMixtureDataset.Component(0.1, [1,1], [[1,0],[0,1]]),
               spn.GaussianMixtureDataset.Component(0.2, [1,4], [[1,0.4],[0.8,1]]),
               spn.GaussianMixtureDataset.Component(0.1, [1,8], [[1,0],[0,1]]),
               spn.GaussianMixtureDataset.Component(0.1, [3,4], [[0.5,0],[0,0.7]]),
               spn.GaussianMixtureDataset.Component(0.1, [3,8], [[1,0],[0,1]]),
               spn.GaussianMixtureDataset.Component(0.1, [5,4], [[1,0],[0,1]]),
               spn.GaussianMixtureDataset.Component(0.3, [5,1], [[1,0.5],[0.5,1]])]
```

```python
dataset1 = spn.GaussianMixtureDataset(components=components1,
                                     num_samples=1000,
                                     num_epochs=1,
                                     batch_size=1000,
                                     shuffle=True)
dataset2 = spn.GaussianMixtureDataset(components=components2,
                                     num_samples=1000,
                                     num_epochs=1,
                                     batch_size=1000,
                                     shuffle=True)
```

# One Class

```python
data = dataset1.get_data()

with spn.session() as (sess, run):
    samples, _, likelihoods = sess.run(data)
    spn.plot_2d(samples[:, 0], samples[:, 1], probs=likelihoods)
```

```python
data = dataset2.get_data()

with spn.session() as (sess, run):
    samples, _, likelihoods = sess.run(data)
    spn.plot_2d(samples[:, 0], samples[:, 1], probs=likelihoods)
```

# Multiple Classes

```python
data = dataset1.get_data()

with spn.session() as (sess, run):
    samples, labels, _ = sess.run(data)
    spn.plot_2d(samples[:, 0], samples[:, 1], labels=labels[:, 0])
```

```python
data = dataset2.get_data()

with spn.session() as (sess, run):
    samples, labels, _ = sess.run(data)
    spn.plot_2d(samples[:, 0], samples[:, 1], labels=labels[:, 0])
```

```python
data = dataset1.get_data()

with spn.session() as (sess, run):
    samples, labels, likelihoods = sess.run(data)
    spn.plot_2d(samples[:, 0], samples[:, 1], labels=labels[:, 0], probs=likelihoods)
```

```python
data = dataset2.get_data()

with spn.session() as (sess, run):
    samples, labels, likelihoods = sess.run(data)
    spn.plot_2d(samples[:, 0], samples[:, 1], labels=labels[:, 0], probs=likelihoods)
```
