```python
import tensorflow as tf
import libspn as spn
```

```python
components = [spn.GaussianMixtureDataset.Component(0.1, [1,1], [[1,0],[0,1]]),
              spn.GaussianMixtureDataset.Component(0.2, [1,4], [[1,0.4],[0.8,1]]),
              spn.GaussianMixtureDataset.Component(0.1, [1,8], [[1,0],[0,1]]),
              spn.GaussianMixtureDataset.Component(0.1, [3,4], [[0.5,0],[0,0.7]]),
              spn.GaussianMixtureDataset.Component(0.1, [3,8], [[1,0],[0,1]]),
              spn.GaussianMixtureDataset.Component(0.1, [5,4], [[1,0],[0,1]]),
              spn.GaussianMixtureDataset.Component(0.3, [5,1], [[1,0.5],[0.5,1]])]
```

# Gaussian Mixture

```python
dataset = spn.GaussianMixtureDataset(components=components,
                                     num_samples=1000,
                                     num_epochs=2,
                                     batch_size=1500,
                                     shuffle=True,
                                     num_threads=1,
                                     allow_smaller_final_batch=True)
```

```python
data = dataset.get_data()

with spn.session() as (sess, run):
    while run():
        samples, labels, likelihoods = sess.run(data)
        spn.plot_2d(samples[:, 0], samples[:, 1], labels=labels[:, 0], probs=likelihoods)
```

# Discrete Gaussian Mixture

```python
dataset = spn.GaussianMixtureDataset(components=components,
                                     num_samples=1000,
                                     num_epochs=2,
                                     batch_size=1500,
                                     shuffle=True,
                                     num_threads=1,
                                     allow_smaller_final_batch=True,
                                     num_vals=20)
```

```python
data = dataset.get_data()

with spn.session() as (sess, run):
    while run():
        samples, labels, likelihoods = sess.run(data)
        spn.plot_2d(samples[:, 0], samples[:, 1], labels=labels[:, 0], probs=likelihoods)
```

# Integer Grid

```python
dataset = spn.IntGridDataset(num_dims=2,
                             num_vals=10,
                             num_epochs=1,
                             batch_size=100,
                             shuffle=True,
                             num_threads=1,
                             allow_smaller_final_batch=True)
```

```python
data = dataset.get_data()

with spn.session() as (sess, run):
    while run():
        samples = sess.run(data)
        spn.plot_2d(samples[:, 0], samples[:, 1], jitter=False)
```
