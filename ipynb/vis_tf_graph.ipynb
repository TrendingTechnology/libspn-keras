```python
import tensorflow as tf
import libspn as spn
```

```python
v1 = spn.IVs(num_vars=1, num_vals=2, name="X1")
v2 = spn.IVs(num_vars=1, num_vals=2, name="X2")
s11 = spn.Sum(v1, name="Sum1.1")
s12 = spn.Sum(v1, name="Sum1.2")
s21 = spn.Sum(v2, name="Sum2.1")
s22 = spn.Sum(v2, name="Sum2.2")
p1 = spn.Product(s11, s21, name="Prod1")
p2 = spn.Product(s11, s22, name="Prod2")
p3 = spn.Product(s12, s22, name="Prod3")
s = spn.Sum(p1, p2, p3, name="Sum")
spn.generate_weights(s)
```

```python
spn.display_spn_graph(s)
```
